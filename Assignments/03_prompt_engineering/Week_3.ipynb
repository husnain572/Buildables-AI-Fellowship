### Question 4
**Prompting with audio to receive a visual representation is called what?**  

**Answer:**  
Prompting with audio to receive a visual representation is called **multimodal prompting**, because the input is provided in one mode (audio) and the output is received in another mode (visual).

### Question 5
**What does a human-in-the-loop approach to prompting emphasize?**  

**Answer:**  
Human-in-the-loop (HITL) refers to a system or process in which a human actively participates in the operation, supervision or decision-making of an automated system.  
In the context of AI, HITL means that humans are involved at some point in the AI workflow to ensure accuracy, safety, accountability or ethical decision-making.  

*Reference:* [IBM - Human in the Loop](https://www.ibm.com/think/topics/human-in-the-loop)

## 1. Reflection Questions

**1.1 How is prompt chaining similar to solving a jigsaw puzzle?**  

***Answer:***  
In a jigsaw puzzle, we match each piece based on the one placed before it. Similarly, in prompt chaining, each new prompt is built on the output of the previous one. In this way, both processes are connected and work in a similar manner.  

---

**1.2 What’s the main difference between *iteration* and *chaining*?**  

***Answer:***  
Iteration involves refining or tweaking a single prompt multiple times to improve the output, while chaining focuses on linking prompts together, where the output of one prompt becomes the input or foundation for the next. In this way, iteration is about enhancing one idea, whereas chaining is about building a sequence of connected ideas to achieve more complex results.  

---

**1.3 Why might prompt chaining be more effective than asking AI one big question?**  

***Answer:***  
LLMs operate within a context window, which allows them to use previous responses as a reference. This makes it easier to design prompts that build on earlier outputs. If we provide a completely new prompt each time, the responses may vary and lack continuity, which might not align with our requirements.  

---

**1.4 How can the *context window* of an AI model affect your chaining strategy?**  

***Answer:***  
The context window has a significant impact on prompt chaining. It allows LLMs to generate more accurate responses based on the history it contains. If the context window is larger, the model can access more information from previous interactions, increasing the chances of producing the desired output.

# 1) Use case — short paragraph, users & benefits, limitations

**Use case (chosen):**
An **IELTS Language Instructor chatbot** that helps students prepare for the IELTS Speaking and Writing modules by correcting grammar, suggesting vocabulary, providing structure for essays, offering sample answers, and giving an approximate band-level estimate with explanation and examples.

**Why I chose this:**
Many IELTS students need fast, consistent feedback on writing and speaking practice. A chatbot gives immediate corrections and model answers so students can iterate an unlimited number of times, which complements classroom study and one-on-one coaching.

**Target users:**

* Self-study IELTS candidates (intermediate → advanced English learners).
* Language centers who want a lightweight practice tool for students.
* Teachers who want a demo/assistant to produce example tasks and corrections.

**Expected benefits:**

* Immediate grammar corrections and targeted vocabulary suggestions.
* Model responses / sample answers for practice.
* Band-estimate plus precise guidance to improve to the next band.
* Scalable 24/7 practice without human tutor cost.

**Limitations (what it cannot do):**

* It cannot *officially* grade or certify scores — band estimates are heuristic, not official.
* It cannot assess spoken pronunciation quality from audio without an integrated STT/TTS pipeline. (Current implementation is text-only.)
* It may hallucinate specifics (e.g., invented references) — so users should verify factual claims.
* It’s not a replacement for a certified tutor for high-stakes advice.

# 2) Justification: Why a chatbot, tasks it performs, and benefits over alternatives

**Why solve this with a chatbot?**

* Speed and interactivity: instant feedback is more effective for iterative language learning.
* Personalization: can provide targeted grammar/vocab suggestions based on user responses.
* Cost-effective scaling: can serve many learners simultaneously.

**How it improves the experience vs website/app/human-only:**

* Compared to static web pages: interactive, reactive, and able to adapt to user content and mistakes.
* Compared to an app without an LLM: can generate custom feedback and examples rather than relying on rigid rules.
* Compared to human support: cheaper and available 24/7 for practice; humans still needed for nuanced coaching and speaking pronunciation.

**Specific tasks the chatbot will perform:**

1. Correct grammar and sentence structure in submitted text.
2. Suggest higher-level vocabulary and collocations with examples.
3. Provide a model answer or model speaking response.
4. Give a band-level estimate (with justification and actionable Next Steps).
5. Provide short drills or mini-tasks to improve a specific weakness.
6. Fallback: answer "I don’t know / refer to resources" for out-of-domain queries.

# 3) Select the LLM (Groq): model choice and justification

**Models Groq supports (examples):** Llama 3 family, Mixtral variants, Gemma family. Groq is optimized for low-latency inference.

**Recommended model for this use case:** `llama-3-8b` (or `mixtral-8x7b` if you prefer Mixtral).
**Why `llama-3-8b`:**

* **Quality:** Good language performance on English tasks (grammar, paraphrase, instructions) while being much less expensive than the 70B models.
* **Latency/cost:** 8B models run faster and cheaper on Groq hardware which is useful for an interactive chat app like this.
* **Real-time UX:** Lower latency improves user experience in Streamlit.
* **Multilingual support:** Decent for English and common languages; if you need specialized non-English support choose a larger model or a multilingual model.

**Tradeoffs:**

* Larger models (e.g., `llama-3-70b`) will likely yield subtlely better feedback and more accurate band estimates, but will cost more and increase latency. For a student-facing MVP, 8B strikes a good balance.

**If you need heavy reasoning and nuance (e.g., high-stakes exam coaching)**, consider `llama-3-70b` but only if budget and latency are acceptable.

# 5) Prompting technique + sample prompt template

**Chosen techniques:**

* **System prompt design** (persona + role constraints).
* **Few-shot prompting** (2 short examples inside system prompt to set style and desired outputs).
* **Instruction + structured output**: ask the model to produce bullet lists and a short JSON-like summary (for consistent UI parsing).

**Why these?**

* System prompt sets consistent persona and evaluation criteria (band estimation style).
* Few-shot helps the model provide consistent formatting (e.g., `Corrections:`, `Vocabulary:`, `Band estimate:`).
* Structured output makes it easier to render in UI and reduces hallucination in style.

# 6) Frontend — Streamlit app 


````python
import streamlit as st
from groq import Groq
import os
from dotenv import load_dotenv

load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")

client = Groq(api_key=GROQ_API_KEY)

SYSTEM_PROMPT = {
    "role": "system",
    "content": (
        "You are an IELTS language instructor. "
        "Help students improve English for IELTS Speaking and Writing. "
        "Correct grammar, suggest vocabulary, and give feedback with band-level estimates. "
        "Always explain with examples."
    )
}

st.set_page_config(page_title="IELTS Instructor Chatbot", page_icon="🎓", layout="centered")

st.title("🎓 IELTS Language Instructor Chatbot")
st.markdown("Ask me IELTS Speaking or Writing questions, and I’ll give you corrections, vocabulary tips, and band estimates.")

if "messages" not in st.session_state:
    st.session_state.messages = [SYSTEM_PROMPT]

if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

user_input = st.chat_input("Type your question here...")

def is_ielts_related(query: str) -> bool:
    keywords = ["ielts", "speaking", "writing", "essay", "grammar", "vocabulary", "band", "exam", "task 1", "task 2"]
    return any(word in query.lower() for word in keywords)

if user_input:
    if not is_ielts_related(user_input):
        reply = "Sorry, I can only answer IELTS-related queries (Speaking/Writing/Grammar/Vocabulary)."
        st.session_state.chat_history.append(("Instructor", reply))
    else:
        st.session_state.messages.append({"role": "user", "content": user_input})
        st.session_state.chat_history.append(("You", user_input))

        st.session_state.messages = [SYSTEM_PROMPT] + st.session_state.messages[-10:]

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=st.session_state.messages,
            max_tokens=512,
            temperature=0.7,
        )
        reply = response.choices[0].message.content

        st.session_state.messages.append({"role": "assistant", "content": reply})
        st.session_state.chat_history.append(("Instructor", reply))

for role, msg in st.session_state.chat_history:
    if role == "You":
        with st.chat_message("user"):
            st.markdown(msg)
    else:
        with st.chat_message("assistant"):
            st.markdown(msg)
````
