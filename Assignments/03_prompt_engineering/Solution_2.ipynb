## NoteBook 2: Design prompts for everyday work tasks

### Exercise 1: Try It Yourself
1. Write down your **to-do list** for today or this week.  
2. Pick **one task** where AI could help.  
   - Example: "Write a follow-up email to my team"  
3. Open your AI tool and **give it this prompt**:  
   - *“Draft a professional, friendly email reminding my team about our weekly meeting.”*  
4. Review, edit, and use the result 

> **AI Prompt**: Draft a professional, friendly email reminding my teammates about our upcoming hackathon project meeting on Thursday at 3 PM, and asking them to share their progress updates before the meeting.

---


### Exercise 2: Try It Yourself
1. Think of an **email you need to write** (e.g., to your team, customers, or a client).  
2. Use the **prompt framework**:  
   - Persona: Who are you?  
   - Context: What’s happening?  
   - Task: What do you want the email to do?  
   - Specificity: Add tone, format, details.  
3. Ask your AI tool to **draft the email**.  
4. Then, **iterate**:  
   - Change the tone (formal → casual, professional → fun).  
   - Adjust the audience.  

> ***Prompt:*** I am the CEO of a software company. As the winter season is approaching, the timetable will change starting next month. The new working hours will be from 8:00 AM to 4:00 PM, with a half-hour break. Write an email to inform all employees about the new timings. The email should have a professional tone and include all necessary details.

### Exercise 3: Try It Yourself
**Practice Tone & Style Prompting**
1. Pick a short text you’ve written (an email, a report, or a social post).  
2. Rewrite your prompt to **specify tone and style**.  
   - Example: “Make this LinkedIn post sound **inspirational, like a TED speaker motivating professionals.**”  
3. Run it through your AI tool.  
4. Iterate: Ask for a new version in a **different tone** (funny, formal, academic, casual).  
5. Compare the results — which version feels most effective for your audience?  

> ***Prompt:*** You are a 3rd-semester university student. The result of your 2nd semester is out, and you achieved a 3.9/4.0 CGPA. Write a LinkedIn post to share with the world that you are doing well and have reached an important milestone. The content of the post should read as if written by a human, not AI. The tone should be natural and not overly excited. The post should inspire HR professionals and students, while also being eye-catching and engaging.
